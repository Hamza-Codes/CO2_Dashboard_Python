```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension ('tabulator') # extension for creating interactive tables

import hvplot.pandas # allows us to create interactive data frames
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv")
df
```

```python
df1 = pd.read_csv("owid-co2-data.csv")
df1
```

```python
df.columns
```

## 1. Some minor data processing  

```python
# Fill N/A with zero (0) and create GDP per capita column

df=df.fillna(0)
df['gdp per capita'] = np.where(df['population']!=0,df['gdp']/df['population'],0) # creates another column, gdp per capita.
# where population is not equal to zero, divide the gdp column by population column
```

```python
df # check the last column - gdp per capita
```

```python
# Make DataFrame pipeline interactive
idf = df.interactive()
```

### CO2 Emissions over time by continent

```python
# Define panel widgets
year_slider = pn.widgets.IntSlider(name="Year Slider",start=1750,end=2020,step=5,value=1850)
year_slider
```

```python
# Radio buttons for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name="Y Axis",
    options=['co2','co2 per capita',],
    button_type = 'success'
)
```

```python
# connect the data pipeline with the widgets, so that as the widgets are changed/interacted with, the underlining data should also change.

idf = df.interactive()
continents = ['World','Asia','Oceania','Europe','Africa','North America','South America','Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country','year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot(x='year',by='country',y=yaxis_co2,line_width=2,title='CO2 emissions by continent')
co2_plot
```

# Table - CO2 Emissions over time by continent

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator,pagination='remote',page_size=10,sizing_mode='stretch_width')
co2_table
```

# CO2 vs GDP Scatter plot

```python
# creating data pipeline for the scatter plot first
co2_vs_gdp_scatterplot_pipeline = (
    idf[(idf.year == year_slider) & (~ (idf.country.isin(continents)))]
    .groupby(['country', 'year', 'gdp per capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

## Plotting the scatter plot

```python
# plotting based on the pipeline data

co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp per capita', 
                                                                y='co2', 
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500)
co2_vs_gdp_scatterplot
```

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='country', 
                                                     y=yaxis_co2_source, 
                                                     title='CO2 source by continent')
co2_source_bar_plot
```
